 ****** Q. 1) Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
cars_df.isnull()
```

```python
cars_df.isna().sum()
```

```python
#second method
null_values = cars_df.isnull().sum()
print("Null Values in Each Column:\n", null_values)
```

```python
# Fill null values with the mean of their respective columns
cars_df.fillna(cars_df.mean(), inplace=True) #this will not work properly b/c there are some cols that contain string values aur string values ka ham mean nhi nikaal sktay





```

```python
#another method
# Find null values in numeric columns
numeric_columns = cars_df.select_dtypes(include=[float, int]).columns
null_numeric_values = cars_df[numeric_columns].isnull().sum()

```

```python
# Fill null values in numeric columns with the mean
for column in numeric_columns:
    if null_numeric_values[column] > 0:
        cars_df[column].fillna(cars_df[column].mean(), inplace=True)
cars_df
```

```python
cars_df.isnull().sum() #again check if there is any col with null values
```

```python
# Find null values in numeric columns
numeric_columns = cars_df.select_dtypes(include=[float, int]).columns
null_numeric_values = cars_df[numeric_columns].isnull().sum()
string_cols = cars_df.select_dtypes(include =[object]).columns 
null_string_values = cars_df[string_cols].isnull().sum()
```

```python
for column in numeric_columns:
    if null_numeric_values[column] > 0:
        cars_df[column].fillna(cars_df[column].mean(), inplace=True)
for column in string_cols:
    if null_string_values[column] > 0:
        cars_df[column].fillna('Unknown', inplace=True)
cars_df
```
