***** Q. 4) Question ( mapping + data-type casting ) 
- What is the mean stop_duration ?

```python
data['stop_duration'].value_counts()
```

```python
#as we notice that dtype of stop_dur is object and we cant find the mean of string vaues so we first convert it into numeric form
data['stop_duration']=data['stop_duration'].map({'0-15 Min' : 7.5,'16-30 Min':24, '30+ Min':45})
```

```python
data
```

```python
#now finally we can find mean duration
data['stop_duration'].mean()
```
